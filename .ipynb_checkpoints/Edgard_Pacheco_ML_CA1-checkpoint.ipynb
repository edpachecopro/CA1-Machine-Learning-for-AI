## Data Loading and Initial Setup
In this section, I imported necessary Python libraries such as Pandas for data manipulation, Matplotlib for data visualization, and scikit-learn for machine learning. The dataset `log2.csv` is loaded into a DataFrame to examine its structure and get initial insights.


```python
import pandas as pd

# Load the dataset
data_path = 'log2.csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataset and its general info
data.info(), data.head()
```


## Exploratory Data Analysis (EDA)
I perform EDA to understand the distributions and relationships in the data. Visualizations help identify patterns and anomalies which are important for selecting appropriate models and predicting the model's performance.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical summary of numeric features
numeric_stats = data.describe()

# Distribution of the "Action" category
action_counts = data['Action'].value_counts()

# Plotting the distribution of "Action"
plt.figure(figsize=(10, 5))
sns.countplot(x='Action', data=data)
plt.title('Distribution of Firewall Actions')
plt.xlabel('Action')
plt.ylabel('Count')
plt.grid(True)

numeric_stats, action_counts

```

```python
# Box plots to check for outliers in key numeric features
plt.figure(figsize=(16, 10))

# List of features to check for outliers
features_to_plot = ['Bytes', 'Bytes Sent', 'Bytes Received', 'Packets', 'pkts_sent', 'pkts_received']

for i, feature in enumerate(features_to_plot):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x=data[feature])
    plt.title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```


## Model Training
4 machine learning models are trained to classify firewall actions. We use a split of training and testing data to evaluate our models. <br> The models include a decision tree and a random forest classifier, Logistic Regression and SVM. This section discusses the training process, parameter tuning, and the rationale behind choosing these models.


## Model Evaluation
The trained models are evaluated using accuracy, precision, and recall metrics. Im analyzing the results to understand the models' strengths and weaknesses in predicting the firewall actions.


```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Assuming 'Action' is the target and other columns are features
X = data.drop('Action', axis=1)
y = data['Action']

# Filter out classes with fewer than 2 instances
filtered_y = y.groupby(y).filter(lambda x: len(x) > 1)
filtered_X = X.loc[filtered_y.index]

# Split data into training and testing sets using stratified sampling
X_train, X_test, y_train, y_test = train_test_split(filtered_X, filtered_y, test_size=0.3, stratify=filtered_y, random_state=42)

# Initialize models
log_reg = LogisticRegression(max_iter=30000)
dec_tree = DecisionTreeClassifier()
rand_forest = RandomForestClassifier()
svm_model = SVC()

# Dictionary of models
models = {
    'Logistic Regression': log_reg,
    'Decision Tree': dec_tree,
    'Random Forest': rand_forest,
    'SVM': svm_model
}

# Train and evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {name}: {accuracy:.2f}")
    print(f"Classification Report for {name}:\n{classification_report(y_test, y_pred, zero_division=1)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Ensure 'Action' is the target and other columns are features
X = data.drop('Action', axis=1)
y = data['Action']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest classifier
random_forest = RandomForestClassifier()

# Train the model
random_forest.fit(X_train, y_train)

# Make predictions on the training data
y_train_pred = random_forest.predict(X_train)

# Make predictions on the testing data
y_test_pred = random_forest.predict(X_test)

# Calculate and print accuracy on training data
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.2f}")

# Calculate and print accuracy on testing data
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Testing Accuracy: {test_accuracy:.2f}")

# Print classification report for testing data
print("Classification Report for Testing Data:")
print(classification_report(y_test, y_test_pred))
```
